```python
import sqlite3

# Classe Animal com m√©todos de salvar e listar
class Animal:
    def __init__(self, nome, especie, som, descricao):
        self.nome = nome
        self.especie = especie
        self.som = som
        self.descricao = descricao

    def salvar_no_banco(self):
        conexao = sqlite3.connect('zoologico.db')
        cursor = conexao.cursor()
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS animais (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                nome TEXT NOT NULL,
                especie TEXT NOT NULL,
                som TEXT NOT NULL,
                descricao TEXT
            )
        ''')
        cursor.execute('''
            INSERT INTO animais (nome, especie, som, descricao)
            VALUES (?, ?, ?, ?)
        ''', (self.nome, self.especie, self.som, self.descricao))
        conexao.commit()
        conexao.close()

    @staticmethod
    def listar_animais():
        conexao = sqlite3.connect('zoologico.db')
        cursor = conexao.cursor()
        cursor.execute('SELECT nome, especie, som, descricao FROM animais')
        animais = cursor.fetchall()
        conexao.close()
        return animais

# Fun√ß√£o para exibir todos os animais do banco
def exibir_animais():
    print("\nüêæAnimais cadastrados no zool√≥gico:\n")
    animais = Animal.listar_animais()
    for nome, especie, som, descricao in animais:
        print(f"üêæNome: {nome}\nEsp√©cie: {especie}\nSom: {som}\nDescri√ß√£o: {descricao}\n" + "-"*50)

# Dados iniciais (somente inseridos se o banco estiver vazio)
def carregar_animais_iniciais():
    conexao = sqlite3.connect('zoologico.db')
    cursor = conexao.cursor()
    cursor.execute('SELECT COUNT(*) FROM sqlite_master WHERE type="table" AND name="animais"')
    if cursor.fetchone()[0] == 0:
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS animais (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                nome TEXT NOT NULL,
                especie TEXT NOT NULL,
                som TEXT NOT NULL,
                descricao TEXT
            )
        ''')
    cursor.execute('SELECT COUNT(*) FROM animais')
    if cursor.fetchone()[0] == 0:
        animais = [
            ("Tigre Siberiano", "Panthera tigris altaica", "Rugido", "Felino majestoso de grande porte."),
            ("Le√£o", "Panthera leo", "Rugido", "O rei da selva."),
            ("Elefante Africano", "Loxodonta africana", "Trombeta", "Comportamento social complexo."),
            ("Baleia Azul", "Balaenoptera musculus", "Canto", "Maior animal do planeta."),
            ("Lobo Cinzento", "Canis lupus", "Uivo", "Ca√ßador inteligente em grupo."),
            ("Orangotango", "Pongo pygmaeus", "Grito", "Primata arbor√≠cola curioso."),
            ("√Åguia Dourada", "Aquila chrysaetos", "Pio agudo", "Ca√ßadora com vis√£o poderosa."),
            ("Urso Polar", "Ursus maritimus", "Rugido", "Habitante do √Årtico."),
            ("Guepardo", "Acinonyx jubatus", "Ronronar", "Animal terrestre mais veloz."),
            ("Gorila", "Gorilla gorilla", "Grunhido", "Primata forte e social."),
            ("Rinoceronte", "Ceratotherium simum", "Bufar", "Segundo maior mam√≠fero terrestre."),
            ("Jacar√©", "Alligator mississippiensis", "Estalo", "Reptil aqu√°tico predador."),
            ("Sucuri-verde", "Eunectes murinus", "Sibilo", "Anaconda da Am√©rica do Sul."),
            ("Leopardo das Neves", "Panthera uncia", "Rugido suave", "Felino do Himalaia."),
            ("Orca", "Orcinus orca", "Assobio", "Golfinho gigante predador."),
            ("Hipop√≥tamo", "Hippopotamus amphibius", "Grunhido", "Mam√≠fero semiaqu√°tico agressivo."),
            ("Lince-ib√©rico", "Lynx pardinus", "Miado agudo", "Felino raro europeu."),
            ("Panda-gigante", "Ailuropoda melanoleuca", "Balido", "Habitante das florestas de bambu."),
            ("Suricato", "Suricata suricatta", "Assobio", "Mam√≠fero social africano."),
            ("Raposa-do-√°rtico", "Vulpes lagopus", "Latido agudo", "Adaptada ao frio."),
            ("Falc√£o-peregrino", "Falco peregrinus", "Chiado", "Ave de rapina veloz."),
            ("Girafa", "Giraffa camelopardalis", "Hum", "Pesco√ßo longo das savanas.")
        ]
        cursor.executemany('''
            INSERT INTO animais (nome, especie, som, descricao)
            VALUES (?, ?, ?, ?)
        ''', animais)
        conexao.commit()
    conexao.close()

# Menu interativo
def menu():
    carregar_animais_iniciais()
    while True:
        print("\n--- Zool√≥gico Interativo ---")
        print("1. Adicionar novo animal")
        print("2. Ver todos os animais")
        print("3. Sair")
        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == '1':
            nome = input("Nome do animal: ")
            especie = input("Esp√©cie do animal: ")
            som = input("Som que ele faz: ")
            descricao = input("Descri√ß√£o do animal: ")
            animal = Animal(nome, especie, som, descricao)
            animal.salvar_no_banco()
            print(f"‚úÖ{nome} foi adicionado!")
        elif opcao == '2':
            exibir_animais()
        elif opcao == '3':
            print("üêí At√© a pr√≥xima visita ao zool√≥gico!")
            break
        else:
            print("Op√ß√£o inv√°lida!")

if __name__ == "__main__":
    menu()
```
